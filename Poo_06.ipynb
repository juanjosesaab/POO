```python
#creamos una clase, definimos la funcion, tienen que estar alineadas.
class Mascota():
 def __init__(self):
    print("Creamos una clase madre")
 def adivina(self):
      print("Hola, soy un perro")
 def hablar(self):
        print("Guaf")
 def correr(self):
  print("*se cansa*")
 def dormir(self):
  print("Zzzzz")
```

```python
# delcaramos la clase madre a una variable
mascota = Mascota()
```

```python
# LLamamos a la respectiva funcion dentro de la clase madre con la variable
mascota.adivina()
```

```python
mascota.correr()
```

```python
mascota.hablar()
```

```python
mascota.dormir()
```

## Crear una herencia

```python
#Creamos una clase heredada con respecto a la clase madre
class Perro(Mascota):

   def __init__(self):
    Mascota.__init__(self)
    print("Creamos una clase heredada")

   def nombre(self):
    print("Luna")

   def idioma(self):
    print("Perrones")

   def raza(self):
    print("Golden")

   def adivina(self):
    print("Se cago")
```

```python
#Llamamos a la clase heredada
perro = Perro()
```

```python
perro.raza()
```

```python
perro.idioma()
```

```python
perro.nombre()
```

```python
#Se llama a la funcion que esta en la clase mencionada
perro.adivina()
```

```python
class Yo():
  def __init__(self):
    print("Me haz invocado")

  def edad(self):
    print("21")

  def nombre(self):
    print("Juan Jos")

  def apodo(self):
    print("JJ")

  def nacionalidad(self):
    print("Ecuatoriana")
```

```python
yo = Yo()
```

```python
yo.nombre()
```

```python
yo.apodo()
```

```python
yo.edad()
```

```python
yo.nacionalidad()
```

```python
class Amigo(Yo):
  def __init__(self):
    Yo.__init__(self)
    print("Hola, soy amigo de JJ")

  def nombre(self):
    print("Alejandro")

  def apodo(self):
    print("Ramirez")

  def edad(self):
    print("20")
```

```python
amigo = Amigo()
```

```python
amigo.nombre()
```

```python
amigo.apodo()
```

```python
amigo.edad()
```
